```python
!pip install -qU lancedb tantivy pdf4llm
import pdf4llm
from google.colab import files
import numpy as np
import lancedb
from lancedb.embeddings import get_registry
from lancedb.pydantic import LanceModel, Vector
from lancedb.rerankers import RRFReranker

# 1️⃣ Upload PDF
uploaded = files.upload()

# Get the uploaded filename
pdf_filename = list(uploaded.keys())[0]



db = lancedb.connect("/content")
# Initialize embedder
#embedder = get_registry().get("sentence-transformers").create()
embedder = get_registry().get("huggingface").create(name='sentence-transformers/all-MiniLM-L6-v2', device="cuda" )
# Define schema
class PDFSchema(LanceModel):
    text: str = embedder.SourceField()
    vector: Vector(embedder.ndims()) = embedder.VectorField()
    page_name: str

# Step 1: Get user input for page selection
page_input = input("Enter pages or ranges (e.g., 1,3-5,7): ")

# Step 2: Convert input into a list of zero-based page indices
selected_pages = []
for part in page_input.split(","):
    part = part.strip()
    if "-" in part:
        start, end = map(int, part.split("-"))
        selected_pages.extend(range(start - 1, end))  # Zero-based index
    else:
        selected_pages.append(int(part) - 1)
selected_page_data = pdf4llm.to_markdown(pdf_filename, page_chunks=True, pages=selected_pages)

# Step 4: Store extracted text and embeddings
entries = []
for page_data in selected_page_data:
    page_name = page_data["metadata"]["file_path"]
    text = page_data["text"]


    entry = {
        "text": text,
        "page_name": page_name
    }
    entries.append(entry)

# Create or overwrite table with schema
tbl = db.create_table("pdf_data", schema=PDFSchema, mode="overwrite")
tbl.add(entries)
```

```python
blooms_model_name = "cip29/blooms_bert"  # Update with your model name if different
blooms_tokenizer = BertTokenizer.from_pretrained(blooms_model_name)
blooms_model = BertForSequenceClassification.from_pretrained(blooms_model_name, num_labels=6).to("cuda")
```

```python
#Function 2: Classify Questions using Bloom’s Taxonomy BERT
def classify_blooms_taxonomy(question):
    inputs = blooms_tokenizer(question, return_tensors="pt", padding=True, truncation=True).to("cuda")
    with torch.no_grad():
        outputs = blooms_model(**inputs)

    probs = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy().flatten()
    bloom_levels = ["BT1 (Remembering)", "BT2 (Understanding)", "BT3 (Applying)",
                    "BT4 (Analyzing)", "BT5 (Evaluating)", "BT6 (Creating)"]

    # Get highest probability Bloom level
    max_index = probs.argmax()
    predicted_level = bloom_levels[max_index]

    # Store probabilities
    probabilities_dict = {bloom_levels[i]: probs[i] for i in range(len(bloom_levels))}

    return predicted_level, probabilities_dict
```

Classifying the generated questions into blooms taxonomy

```python
import csv
import os
from google.colab import files

def classify_questions_and_save(input_csv="generated_questions.csv", output_csv="classified_questions.csv"):
    local_path = f"/content/{output_csv}"  # Save output in Colab workspace

    # Check if the input file exists
    if not os.path.exists(input_csv):
        print(f"❌ Error: '{input_csv}' not found. Please run `bulk_process_pages()` first.")
        return

    with open(input_csv, mode='r', encoding='utf-8') as infile, open(local_path, mode='w', newline='', encoding='utf-8') as outfile:
        reader = csv.DictReader(infile)  # Read CSV as dictionary to access column names
        writer = csv.writer(outfile)

        # Verify that the column "Question" exists
        if "Question" not in reader.fieldnames:
            print("❌ Error: The column 'Question' is missing in the CSV file.")
            return

        # Write new header for classified questions
        writer.writerow(["Question", "Bloom's Taxonomy Level",
                         "BT1 (Remembering)", "BT2 (Understanding)", "BT3 (Applying)",
                         "BT4 (Analyzing)", "BT5 (Evaluating)", "BT6 (Creating)"])

        for row in reader:
            question = row["Question"].strip()  # Extract question text
            if not question:  # Skip empty questions
                continue

            # Classify the question using the fine-tuned BERT model
            bloom_level, probs = classify_blooms_taxonomy(question)

            writer.writerow([question, bloom_level,
                             probs.get("BT1 (Remembering)", 0), probs.get("BT2 (Understanding)", 0),
                             probs.get("BT3 (Applying)", 0), probs.get("BT4 (Analyzing)", 0),
                             probs.get("BT5 (Evaluating)", 0), probs.get("BT6 (Creating)", 0)])

    print(f"✅ Blooms taxonomy Classified questions saved in {local_path}")
    files.download(local_path)  # Allow user to download the file

# Run the function after generating questions
classify_questions_and_save()
```

```python
import os
import csv

def display_questions_by_bloom_level(input_csv="/content/classified_questions.csv"):
    if not os.path.exists(input_csv):  # Check if the file exists in the Colab environment
        print(f"❌ Error: The file '{input_csv}' was not found. Ensure the classification step was completed.")
        return

    bloom_levels = {
        "1": "BT1 (Remembering)",
        "2": "BT2 (Understanding)",
        "3": "BT3 (Applying)",
        "4": "BT4 (Analyzing)",
        "5": "BT5 (Evaluating)",
        "6": "BT6 (Creating)"
    }

    print("\nSelect Bloom’s Taxonomy Level:")
    print("1: Remembering, 2: Understanding, 3: Applying, 4: Analyzing, 5: Evaluating, 6: Creating")

    bloom_choice = input("Enter the number corresponding to the level: ").strip()
    num_questions = int(input("Enter the number of questions to display: ").strip())

    if bloom_choice not in bloom_levels:
        print("Invalid choice! Please enter a number between 1 and 6.")
        return

    selected_level = bloom_levels[bloom_choice]
    filtered_questions = []

    # Read the classified questions from CSV
    with open(input_csv, mode='r', encoding='utf-8') as infile:
        reader = csv.reader(infile)
        next(reader)  # Skip header row

        for row in reader:
            if len(row) < 3:  # Ensure there are at least three columns
                continue

            question, bloom_level = row[:2]
            if bloom_level == selected_level:
                filtered_questions.append(question)  # Append only the question

    print(f"\nDisplaying {min(num_questions, len(filtered_questions))} questions from {selected_level}:\n")

    for i, question in enumerate(filtered_questions[:num_questions], 1):  # Fixed unpacking issue
        print(f"{i}. {question}")  # Removed page reference

    if not filtered_questions:
        print(f"No questions found for {selected_level}.")

# Run the function after saving the CSV
display_questions_by_bloom_level()
```
