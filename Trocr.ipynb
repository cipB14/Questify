```python
# Step 1: Upload the image
from google.colab import files

uploaded = files.upload()

# Step 2: Load the uploaded image
from PIL import Image
import io

# Get the uploaded file name
uploaded_file_name = next(iter(uploaded))

# Load the image
image = Image.open(io.BytesIO(uploaded[uploaded_file_name])).convert("RGB")

# Step 3: Load the pipeline
from transformers import pipeline

# Load the image-to-text pipeline
pipe = pipeline("image-to-text", model="microsoft/trocr-base-handwritten")

# Step 4: Extract text from the image
result = pipe(image)

# Step 5: Print the result
print(result)
```
